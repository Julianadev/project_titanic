## Documentação do Projeto Titanic
Este projeto analisa dados do Titanic a partir do arquivo titanic.csv. O objetivo é extrair informações sobre passageiras do sexo feminino da primeira classe que sobreviveram ao acidente do Titanic e gerar um novo arquivo CSV com os resultados.

## Passos:

Importação das bibliotecas:

Importa as bibliotecas pandas para manipulação de dados.

## Leitura dos dados:

Lê o arquivo titanic.csv em um DataFrame df.

## Criação de um novo DataFrame:

Cria um novo DataFrame df_novo com as colunas 'Name', 'Age', 'Sex', 'Pclass' e 'Survived' a partir do DataFrame original df.

## Filtragem dos dados:

Filtra o DataFrame df_novo para obter apenas passageiras do sexo feminino da primeira classe, armazenando o resultado em filter1.

## Ordenação dos dados:

Ordena o DataFrame filter1 em ordem alfabética pelo nome, armazenando o resultado em order_asc.

## Exportação dos resultados:

Exporta o DataFrame order_asc para um arquivo CSV chamado titanic_result.csv.

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
columns = {
    'Name': df['Name'],
    'Age': df['Age'],
    'Sex': df['Sex'],
    'Pclass': df['Pclass'],
    'Survived': df['Survived']
}
```

```python
df_novo = pd.DataFrame(columns)

filter1 = df_novo.loc[(df_novo['Sex'] == 'female') & (df_novo['Pclass'] == 1) & (df_novo['Survived'] == 1)]
filter1
```

```python
order_asc = filter1.sort_values(['Name'], ascending=True)
order_asc
```

```python
order_asc.to_csv('titanic_result.csv', index=False)
```
